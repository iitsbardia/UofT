# Transcendental functions

### What is a function?

A function consists of

$$
\begin{array}{c}
\text{1. a domain}\quad A \\
\text{2. a co-domain}\quad B \\
\text{3. a rule that matches each input}\quad x\in A \quad\text{to exactly one output}\quad f(x) \in B
\end{array}
$$
$$
f: A \rightarrow B
$$

Ex : $f: \mathbb{R} \rightarrow \mathbb{R}$, $f(x) = x^2 + \sin x + 3$

Range : $\text{Range} f$ = {$f(a)$ | $a \in A$}  
In calculus, we assume all functions we define are **Surjective** (the **range** and **co-domain** are the same)

### Inverse functions 

(a function is **one-to-one** or **injective** when $\forall x_{1},x_{2}, \quad x_{1}\neq x_{2} \implies f(x_{1}) \neq f(x_{2})$)

Let $f$ be a one-to-one function $s.t.$
$$
Domain f  = A \qquad Range f = C\\
$$
The inverse function $f^{-1}$ is another function satisfying
$$
Domain f^{-1}  = C \qquad Range f^{-1} = A\\
$$
And defined
$$
\forall x\in A,\forall y\in C \qquad y = f(x) \iff x = f^{-1}(y)
$$

$\forall x\in A,\forall y\in C$
1. $f^{-1}(f(x)) = x$
2. $f(f^{-1}(y)) = y$

###  Derivative of the inverse of a function

let $f$ be a function on $I$

IF
- $f$ has an inverse
- $f$ is differentiable
- $\forall x\in I, \quad f'(x) \neq 0$

THEN
- $f^{-1}$ is differentiable
$$
(f^{-1})'(x) = \frac{1}{f'(f^{-1}(x))}
$$

### -$f(x) = a^x$, What is $f'(x)$ ?

$$
f'(x) = \lim_{h \to 0} \frac{a^{x+h} - a^x}{h} = \lim_{h \to 0} \frac {a^x (a^h - 1)}{h} = a^x .\lim_{h \to 0} \frac{a^h - 1}{h}
$$
($\lim_{h \to 0} \frac{a^h - 1}{h}$ = $L_{a}$)
$$
\frac{d}{dx} a^x = L_{a} a^x
$$

## lets define $e$

$e$ is the only number where $L_{e}$ = 1
$$
\frac{d}{dx} e^x = e^x
$$

---

### -$\frac{d}{dx} e^x = e^x$, What is $\frac{d}{dx} \ln x$ ?

$$
e^{\ln x} = x \implies \frac{d}{dx} [e^{\ln x}] = \frac{d}{dx} [x] \implies \frac{d}{dx} [e^{\ln x}] = 1
$$
$$
\implies e^{\ln x}.\frac{d}{dx}[\ln x] = 1 \implies \mathbf{\frac{d}{dx}[\ln x]} = \frac{1}{e^{\ln x}} = \mathbf{\frac{1}{x}}
$$

---

### -$\frac{d}{dx} e^x = e^x$, What is $\frac{d}{dx} a^x$ ?

$$
a^x = e^{x\ln a} \implies \mathbf{\frac{d}{dx} [a^x]} = \frac{d}{dx} [e^{x\ln a}] = e^{x\ln a}.\frac{d}{dx}[x\ln a] = a^x.\ln a.\frac{d}{dx}x = \mathbf{a^x \ln a}
$$

---

### -$\frac{d}{dx} \ln x = \frac{1}{x}$, What is $\frac{d}{dx} \log_{a} x$ ?

$$
\log_{a} x = \frac{\ln x}{\ln a} \implies \mathbf{\frac{d}{dx}[\log_{a} x]} = \frac{d}{dx}[\frac{\ln x}{\ln a}] = \frac{1}{\ln a}.\frac{d}{dx}[\ln x] = \mathbf{\frac{1}{(\ln a) (x)}}
$$

---

### Logarithmic Differentiation $\frac{d}{dx} [f(x)^{g(x)}] = ?$

Ex: $\frac{d}{dx} [cos(x)^{sin(x)}]$
$$
cos(x)^{sin(x)} = e^{sin(x) \ln cos(x)} \implies \mathbf{\frac{d}{dx} [cos(x)^{sin(x)}]} = \frac{d}{dx} e^{sin(x) \ln cos(x)} = e^{sin(x) \ln cos(x)}. \frac{d}{dx} [sin(x) \ln cos(x)]
$$
$$
= cos(x)^{sin(x)}.[cos(x).\ln cos(x) + sin(x).\frac{1}{cos(x).-sin(x)}] = \mathbf{cos(x)^{sin(x)}.[cos(x).\ln cos(x)-\frac{sin^2(x)}{cos(x)}]}
$$

---

### Proof of the power rule (for all exponents)

$$
\frac{d}{dx} x^C = \frac{d}{dx} e^{c \ln x} = e^{c \ln x}.\frac{d}{dx}[c \ln x] = x^c.c.\frac{1}{x} = c.x^{c-1}
$$

## ARCSIN

$$
\mathbf{\arcsin} \text{ is the inverse function of } \mathbf{\sin}
$$
$$
\text{restricted to} [-\frac{\pi}{2},\frac{\pi}{2}]
$$

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi/2, np.pi/2, 400)
y_sin = np.sin(x)
x_arcsin = np.linspace(-1, 1, 400)
y_arcsin = np.arcsin(x_arcsin)
plt.figure(figsize=(10, 6))
plt.plot(x, y_sin, label=r'$\sin(x)$', color='blue')
plt.plot(x_arcsin, y_arcsin, label=r'$\arcsin(x)$', color='red')
plt.title(r'Plot of $\sin(x)$ and $\arcsin(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend()
plt.grid(True)
plt.show()
```

$$
\forall x\in[-\frac{\pi}{2},\frac{\pi}{2}],\forall y\in[-1,1]
$$
$$
x = \arcsin y \iff y = \sin x
$$

Ex : what is $\arcsin \frac{1}{2}$ ?

$\sin (\arcsin\frac{1}{2}) = \frac{1}{2} $ (since $\sin \frac{\pi}{6} = \frac{1}{2}$) $\implies$ $\arcsin \frac{1}{2} = \frac{\pi}{6}$

#### $\frac{d}{dx} \arcsin x$ 

$x\in [-1,1] \sin(\arcsin (x)) = x$, thus
$$
\frac{d}{dx} \sin(\arcsin (x)) = \frac{d}{dx} x \implies \cos(\arcsin (x)).\frac{d}{dx}[\arcsin (x)] = 1 \implies \frac{d}{dx} \arcsin (x) = \frac{1}{\cos(\arcsin (x))} = \frac{1}{\sqrt{1-x^2}}
$$

## ARCTAN

$$
\mathbf{\arctan} \text{ is the inverse function of } \mathbf{\tan}
$$
$$
\text{restricted to} (-\frac{\pi}{2},\frac{\pi}{2})
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x_tan = np.linspace(-np.pi/2 + 0.01, np.pi/2 - 0.01, 400)
y_tan = np.tan(x_tan)
x_arctan = np.linspace(-10, 10, 400)
y_arctan = np.arctan(x_arctan)
plt.figure(figsize=(10, 6))
plt.plot(x_tan, y_tan, label=r'$\tan(x)$', color='blue')
plt.plot(x_arctan, y_arctan, label=r'$\arctan(x)$', color='red')
plt.title(r'Plot of $\tan(x)$ and $\arctan(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.xlim(-7, 7)
plt.ylim(-5, 5)
plt.legend()
plt.grid(True)
plt.show()
```

$$
\forall x\in(-\frac{\pi}{2},\frac{\pi}{2}),\forall y\in\mathbb{R}
$$
$$
x = \arctan y \iff y = \tan x
$$

## ARCCOS

$$
\mathbf{\arccos} \text{ is the inverse function of } \mathbf{\cos}
$$
$$
\text{restricted to} [0,{\pi}]
$$

$$
\forall x\in[0,\pi],\forall y\in[-1,1]
$$
$$
x = \arccos y \iff y = \cos x
$$

---

$$
\begin{array}{c}
\ \frac{d}{dx}[\arcsin (x)] = \frac{1}{\sqrt{1-x^2}} \\
\ \frac{d}{dx}[\arccos (x)] = -\frac{1}{\sqrt{1-x^2}} \\
\ \frac{d}{dx}[\arctan (x)] = \frac{1}{1+x^2}
\end{array}
$$
