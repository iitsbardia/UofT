# The formal definition of limit

#### limit at a point

1. $x$ is close to $a$ and not $a$
$$
    0 < \mid x - 1 \mid < \delta
$$
1. **IF** $x$ is close to $a$ and not $a$, **THEN** $f(x)$ is close to $L$
$$
0 < \mid x - 1 \mid < \delta \Rightarrow \mid f(x) - L \mid < \varepsilon
$$

but this on its own is meaningless, to have a formal definition we need to define our variables:
* let $a$ , $L$ $\in \mathbb{R}$
* let $f(x)$ be a function on an interval, at least around $a$ except maybe at $a$
* We say that
$$ 
\lim_{x\to a}f(x) = L
$$
* When
$$
\forall \varepsilon > 0 ,\exists \delta >0 s.t. (\forall x \in \mathbb{R})
$$
$$
0<\mid x - a \mid < \delta \Rightarrow \mid f(x) - L \mid < \varepsilon
$$

#### Limit at infinity

so we defined a limit at a point but what about $\infty$. well to define a limit when x aproaches infinity we need to show
* let $M$ , $L$ $\in \mathbb{R}$
* let $f(x)$ be a function on an interval$(p,\infty)$ (for some $p \in \mathbb{R}$ )
* We say that
$$ 
\lim_{x\to \infty}f(x) = L
$$
* When
$$
\forall \varepsilon > 0 ,\exists M >0 s.t. (\forall x \in \mathbb{R})
$$
$$
x > M  \Rightarrow \mid f(x) - L \mid < \varepsilon
$$

## Proving limits from definition

So now that we know the definition of limit let's try some examples

Ex1: Prove $\lim_{x\to 3}f(2x + 1) = 7$ directly from definition

pf:
* let $a$ , $L$ $\in \mathbb{R}$
* let $\varepsilon$ > 0
* take $\delta$ = $\frac{\varepsilon}{2}$
* Let $x\in\mathbb{R}$ , Assume $0<\mid x - 3 \mid < \delta$, we need to confirm  $\mid 2x+1 - 7 \mid < \varepsilon$
* Since $\mid x - 3 \mid < \delta$ = $\frac{\varepsilon}{2}$, Then $2 \mid x- 3 \mid$ = $\mid 2x- 6 \mid$ < $\varepsilon$
* we have shown $\mid 2x+1 - 7 \mid < \varepsilon$, as needed $\square$

---

Ex2: Prove $\lim_{x\to 4}f(x^2 + 1) = 17$ directly from definition

pf:
* let $a$ , $L$ $\in \mathbb{R}$
* let $\varepsilon$ > 0
* take $\delta$ = $min({1},{\frac{\varepsilon}{9}})$
* Let $x\in\mathbb{R}$ , Assume $0<\mid x - 4 \mid < \delta$, we need to confirm  $\mid x^2+1 - 17 \mid < \varepsilon$
* Since $\mid x - 4 \mid < 1$, Then $\mid x + 4\mid < 9$
* And by assumption $\mid x - 4 \mid < \frac{\varepsilon}{9}$
* thus, $\mid x^2+1 - 17 \mid = \mid x - 4 \mid\mid x + 4 \mid  < 9\frac{\varepsilon}{9}= \varepsilon$
* We have shown $\mid x^2+1 - 17 \mid < \varepsilon$, as needed $\square$

## Limit Laws

Assume
$$
\lim_{x\to a}f(x) = L and \lim_{x\to a}g(x) = M
$$  
Then
$$
\lim_{x\to a}[f(x)+g(x)] = L + M
$$
$$
\lim_{x\to a}[f(x).g(x)] = L.M
$$
Note: **Limit Laws only apply if the initial limit exists!**

####  Proof of the limit law for sums

---

Thm:
* Let $f and g$ be functions centred around $a$ except maybe at a
* Let $L,M \in\mathbb{R}$

IF
$$
\lim_{x\to a}f(x) = L and \lim_{x\to a}g(x) = M
$$
Then
$$
\lim_{x\to a}[f(x)+g(x)] = L + M
$$

---

WTS:
$$
\forall\varepsilon >0 , \exists \delta > 0 s.t
$$
$$
0<\mid x - a \mid<\delta \Rightarrow \mid [f(x) + g(x)] - [L+M] \mid < \varepsilon
$$

---

pf:
* Let $\varepsilon > 0$
* use $\frac{\varepsilon}{2}$ in definition of $\lim_{x\to a}f(x) = L$
* we get $\exists \delta_{1} > 0 s.t$ $0<\mid x - a \mid<\delta_{1} \Rightarrow \mid f(x) - L \mid < \frac{\varepsilon}{2}$
* use $\frac{\varepsilon}{2}$ in definition of $\lim_{x\to a}g(x) = M$
* we get $\exists \delta_{1} > 0 s.t$ $0<\mid x - a \mid<\delta_{2} \Rightarrow \mid g(x) - M \mid < \frac{\varepsilon}{2}$
* take $\delta$ = $min(\delta_{1},\delta_{2})$
* Let $x\in \mathbb{R}$, Assume $0<\mid x - a \mid<\delta$
* Since $\delta \leq \delta_{1}$ ,thus $\mid f(x) - L \mid < \frac{\varepsilon}{2}$
* Since $\delta \leq \delta_{2}$ ,thus $\mid g(x) - M \mid < \frac{\varepsilon}{2}$
* if we add the two statements we get $\mid [f(x) + g(x)] - [L+M] \mid < \varepsilon$, as needed $\square$

## The Squeeze Theorem

The Squeeze Theorem states that if a function $f(x)$ is squeezed between two other functions $g(x),h(x)$ whose limits are equal as $x$ approaches a point then the limit of $f(x)$ as $x$ approaches that point is also the same.

##### To write a rigorous proof:

- let $L,a\in\mathbb{R}$
- let $f,g,h$ be function defined near $a$ except maybe at $a$

IF
* $\exists p>0 s.t. 0<\mid x-a \mid<p$ $\Rightarrow$ $h(x) \leq f(x) \leq g(x)$
* $\lim_{x\to a}h(x) = \lim_{x\to a}g(x) = L$

Then
* $\lim_{x\to a}f(x) = L$

```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    return -x**2

def f(x):
    return x**2 * np.sin(1/x)

def h(x):
    return x**2

x = np.linspace(-0.1, 0.1, 1000)
x = x[np.abs(x) > 1e-4]
g_values = g(x)
f_values = f(x)
h_values = h(x)
plt.figure(figsize=(10, 6))
plt.plot(x, g_values, label='$-x^2$', color='red')
plt.plot(x, f_values, label='$x^2 \\sin(1/x)$', color='blue')
plt.plot(x, h_values, label='$x^2$', color='green')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.ylim(-0.02, 0.02)
plt.xlim(-0.1, 0.1)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Visualization of the Squeeze Theorem')
plt.legend()
plt.grid(True)
plt.show()
```

#### Proof of the Squeeze Theorem

* let $\varepsilon \in\mathbb{R}$
* use $\varepsilon$ in definition of $h(x)$ , $\exists \delta_{1} s.t.$  $0<\mid x-a \mid<\delta_{1}$ $\Rightarrow$ $\mid h(x)-L\mid<\varepsilon$
* use $\varepsilon$ in definition of $g(x)$ , $\exists \delta_{2} s.t.$ $0<\mid x-a \mid<\delta_{2}$ $\Rightarrow$ $\mid g(x)-L\mid<\varepsilon$
* Take $\delta = min(\delta_{1},\delta_{2},p)$
* Let $x\in \mathbb{R}$, Assume $0 < \mid x-a \mid < \delta$, this implies
    - $ \delta  \leq \delta _{1}$, thus $\mid h(x)-L\mid<\delta$
    - $ \delta  \leq \delta _{2}$, thus $\mid g(x)-L\mid<\delta$
    - $ \delta  \leq p$, thus $h(x) \leq f(x) \leq g(x)$
* we have shown  $\mid f(x)-L\mid<\varepsilon$, as needed $\square$
