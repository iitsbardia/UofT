## The definition of continuity

**Definition**

* Let $a\in\mathbb{R}$
* Let $f$ be a function on an interval, at least centered around$a$

We say $f$ is continuous when
$$
\lim_{x\to a}f(x) = f(a)
$$

This means:
1. $\lim_{x\to a}f(x)$ exists
2. $f(a)$ is defined
3. $\lim_{x\to a}f(x) = f(a)$

**The $\delta$,$\varepsilon$ Definition**
$$
\forall \varepsilon > 0 , \exists \delta >0 s.t.
$$
$$
\mid x - a \mid < \delta \Rightarrow \mid f(x) - f(a) \mid < \varepsilon
$$

---
- continuous at a point
$$
\lim_{x\to c}f(x) = f(c)
$$
- continuous on $(a.b)$
$$
\forall c \in (a,b),\lim_{x\to c}f(x) = f(c)
$$
- continuous on $[a,b]$
$$
\lim_{x\to a^{-}}f(x) = f(a)
$$
$$
\forall c \in (a,b),\lim_{x\to c}f(x) = f(c)
$$
$$
\lim_{x\to b^{+}}f(x) = f(b)
$$

### The main continuity theorem

Theorem:  polynomial, rational, root, trigonometric, absolut values, exponential, and logarithmic functions are continuous at every number **"in their domain"**.

How to Prove it

### Discontinuities

1. Removeable Discontinuities

$\lim_{x\to a}f(x)$ exists, but $\lim_{x\to a}f(x) \neq f(a)$

2. None Removeable Discontinuities

$\lim_{x\to a}f(x)$ $DNE$

#### Removeable Discontinuities can be fixed

Ex: $h(x) = \frac{\sin x}{x}$, $\lim_{x \to 0}h(x) = 1$, but $h(0)$ Undifined

we can define a new function
$$
H(x) =
\begin{cases} 
\frac{\sin x}{x} & \text{if } x \neq 0 \\
1 & \text{if } x = 0 
\end{cases}
$$

Then $H$ continuous at 0

```python
import numpy as np
import matplotlib.pyplot as plt

def H(x):
    return np.where(x != 0, np.sin(x) / x, 1)
def h(x):
    return np.sin(x) / x

x = np.linspace(-10, 10, 400)
x_h = x[x != 0]
y_H = H(x)
y_h = h(x_h)
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(x, y_H, label='$H(x)$', color='blue')
plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
plt.title('$H(x)$')
plt.xlabel('$x$')
plt.ylabel('$H(x)$')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(x_h, y_h, label='$h(x)$', color='red')
plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
plt.scatter(0, 1, color='red', zorder=5, label='Undefined at $x=0$')  # Mark undefined point
plt.title('$h(x)$')
plt.xlabel('$x$')
plt.ylabel('$h(x)$')
plt.legend()
plt.tight_layout()
plt.show()
```

Problem: prove $\lim_{x\to 0}\frac{\sin x}{x} = 1$

* we can geometrically prove that $\cos x \leq \frac{\sin x}{x} \leq \frac{1}{\cos x}$
* and we know $\lim_{x\to 0} \cos x = \lim_{x\to 0} \frac{1}{\cos x} = 1$
* Hence, by the squeeze theorem we get

$$
\lim_{x\to 0}\frac{\sin x}{x} = 1
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def sin_x_over_x(x):
    return np.sin(x) / x
def cos_x(x):
    return np.cos(x)
def one_over_cos_x(x):
    return 1 / np.cos(x)

x = np.linspace(-1, 1, 1000)
x_nonzero = x[x != 0]
y_sin_x_over_x = sin_x_over_x(x_nonzero)
y_cos_x = cos_x(x)
y_one_over_cos_x = one_over_cos_x(x)
plt.figure(figsize=(10, 6))
plt.plot(x_nonzero, y_sin_x_over_x, label=r'$\frac{\sin(x)}{x}$', color='blue')
plt.plot(x, y_cos_x, label=r'$\cos(x)$', color='green', linestyle='--')
plt.plot(x, y_one_over_cos_x, label=r'$\frac{1}{\cos(x)}$', color='red', linestyle='--')
plt.scatter(0, 1, color='blue')
plt.axhline(1, color='black', linestyle='--', linewidth=0.5)
plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
plt.title('Visualization of the Squeeze Theorem')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.ylim(-2, 2)
plt.legend()
plt.grid(True)
plt.show()
```

# EVT, IVT

---

### Extreme Value Theorem

IF

$f$ is a continuous function on an interval $[a,b]$.

THEN

$f$ has a max and min on $[a,b]$.

---

### Intermediate Value Theorem

IF

* $f$ is a function defined on an interval $[a,b]$
* $f$ is continuous on $[a,b]$
* $M \in \mathbb{R}$
$$
f(a) > M > f(b)
$$

THEN

$f$ has a max and min on $[a,b]$.
$$
\exists c \in [a,b] s.t. f(c) = M
$$
