# Unit 2: Limits and Continuity

#### The idea of limit
$$
\lim_{x\to a} f(x) = L
$$
This statement roughly means that when **x is very close to a** the **f(x) is very close to L**.(note that this is not the formal definition of limit but just an idea of what limit means)

Let's see an example:
* $f(x) = \frac{x^2 - 1}{x-1}$
* f(1) is undefined
* (if $x \neq 1$) $\frac{x^2 - 1}{x-1}$ = $\frac{(x-1)(x+1)}{x-1}$ = $x+1$
* so we can define a new function $g(x) = x+1$  
As we can see when $x$ is close to $1$ $f(x)$ is close to $2$

```python
import matplotlib.pyplot as plt
import numpy as np


def f(x):
    return np.where(x != 1, (x**2 - 1) / (x - 1), np.nan)
def g(x):
    return x + 1

x = np.linspace(-2, 2, 400)
x_f = np.linspace(-2, 2, 400)
x_f = x_f[x_f != 1]
y_f = f(x_f)
y_g = g(x)
fig, axs = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns
axs[0].plot(x_f, y_f, color='blue', label='f(x) = (x^2 - 1) / (x - 1)')
axs[0].scatter([1], [2], color='red', zorder=5)  # Show where f(x) is undefined
axs[0].annotate('Undefined\nat x=1', xy=(1, 2), xytext=(1.5, 2.5),
                arrowprops=dict(facecolor='black', shrink=0.05),
                fontsize=10, color='red')
axs[0].set_title('f(x) = (x^2 - 1) / (x - 1)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)
axs[1].plot(x, y_g, color='green', label='g(x) = x + 1')
axs[1].set_title('g(x) = x + 1')
axs[1].set_xlabel('x')
axs[1].set_ylabel('g(x)')
axs[1].grid(True)
axs[0].legend()
axs[1].legend()
plt.tight_layout()
plt.show()
```

### Examples of limits that do not exist

So now lets Try and find limit of $h(x) = \sin{\frac{\pi}{2x}}$ as $x$ aproaches $0$

As you can see no as $x$ approaches $0$,$h(x)$ isn't close to any number it is all over the place between $-1$ and $1$ here we say
$$
\lim_{x\to0}h(x) DNE
$$

```python
import numpy as np
import matplotlib.pyplot as plt


def h(x):
    return np.sin(np.pi / (2 * x))


x = np.linspace(-10, 10, 1000)
x = x[x != 0]
y = h(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$h(x) = \sin{\left(\frac{\pi}{2x}\right)}$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title(r'Plot of $h(x) = \sin{\left(\frac{\pi}{2x}\right)}$')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.legend()
plt.show()
```

Lets Try another one, find limit of $f(x) = \frac{1}{(x-1)^2}$ as $x$ aproaches $1$

As you can see no as $x$ approaches $1$,$f(x)$ isn't close to any number it is getting aribtrary large,here we say
$$
\lim_{x\to1}f(x) = \infty
$$
This Limit "**Does not exist**"


```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return 1 / (x - 1)**2

x = np.linspace(0.5, 1.5, 1000)
x = x[x != 1]
y = f(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x) = \frac{1}{(x-1)^2}$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(1, color='red', linestyle='--', linewidth=0.5, label='x = 1')
plt.ylim(0, 100000)  # Limit y-axis to make the plot more readable
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title(r'Plot of $f(x) = \frac{1}{(x-1)^2}$ near $x = 1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

### Side Limits

lets Try and find limit of $g(x) = \frac{x^2+X}{\mid{x}\mid}$ as $x$ aproaches $0$

As you can see no as $x$ approaches $0$,$g(x)$ isn't close to any number but if we only take the right side its getting close to $1$ and if we only take the left side its getting close to $-1$
$$
\lim_{x\to0}G(x) DNE
$$
$$
\lim_{x\to0^{+}}G(x) = 1
$$
$$
\lim_{x\to0^{-}}G(x) = -1
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    return (x**2 + x) / np.abs(x)

x_values_left = np.linspace(-1, -0.01, 500)
x_values_right = np.linspace(0.01, 1, 500)
y_values_left = g(x_values_left)
y_values_right = g(x_values_right)
plt.figure(figsize=(10, 6))
plt.plot(x_values_left, y_values_left, label=r'$g(x) = \frac{x^2 + x}{|x|}$ (x < 0)', color='blue')
plt.plot(x_values_right, y_values_right, label=r'$g(x) = \frac{x^2 + x}{|x|}$ (x > 0)', color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Plot of $g(x) = \\frac{x^2 + x}{|x|}$ around $x = 0$')
plt.xlabel('$x$')
plt.ylabel('$g(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
