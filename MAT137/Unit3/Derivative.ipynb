# Derivative

### Derivative as slope

* Let $f$ be function on $I$
* Let $a\in I$

**$f'(a)$** = Slope of line tangent to point $a$ on the graph of $y = f(x)$

we can calculate the slope of the line from 2 points $a$ and $b$ on the tangent line as 
$$
\frac{f(a)-f(b)}{a-b}
$$
but what about at a single point on a function like $a$, we can use limit to calculate that
$$
f'(a) = \lim_{x\to a} \frac{f(x)-f(a)}{x-a}
$$
which also can be written as
$$
f'(a) = \lim_{x\to a} \frac{f(x)-f(a)}{x-a} = \lim_{h\to 0} \frac{f(a+h)-f(a)}{h}
$$

and by calcluating $f'(a)$ we can write the tanget line as 
$$
y = f'(a)(x - a) + f(a)
$$

---

Ex: compute $f'(1)$ for the function $f(x) = 4x - x^2$

$$
f'(1) = \lim_{x\to 1} \frac{f(x)-f(1)}{x-1} = \lim_{x\to 1} \frac{4x - x^2-(4-1)}{x-1} = \lim_{x\to 1} \frac{(x-1)(3-x)}{x-1} =(x \neq 1) \lim_{x\to 1} 3-x = 3-1 = 2
$$

---

### Rate of change
$$
\frac{dx}{dt} = \lim_{\Delta t \to 0} \frac{\Delta x}{\Delta t} = \lim_{x\to a} \frac{f(x)-f(a)}{x-a} = f'(a)
$$

---

### Differentiable $\Rightarrow$ Continuous

pf:

* Assume f is differentiable at $c$
* Then:

$$
\lim_{x\to c}[f(x) - f(c)] =(x \neq c) \lim_{x\to c} [\frac{f(x)-f(c)}{x-c}.(x-c)] = \lim_{x\to c}[f(x)-f(c)] \lim_{x \to c} (x-c) = f'(c) . 0 = 0 
$$

* Thus we have shown $\lim_{x\to c}[f(x) - f(c)] = 0$, as needed $\square$.

remember : $f$ is Continuous at c: $\lim_{x\to c}f(x) = f(c)$, its the same as $\lim_{x\to c}[f(x) - f(c)] = 0$

### Differentiable $\nLeftarrow$ Continuous

As you can see both $f(x) = \mid x \mid$ and $f(x) = x^{\frac{1}{3}}$ are Continuous at x = 0 but neither are Differentiable

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return np.abs(x)
def f2(x):
    return np.cbrt(x)
x = np.linspace(-2, 2, 400)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.plot(x, f1(x), label=r'$f(x) = |x|$', color='blue')
ax1.scatter([0], [f1(0)], color='green', zorder=5)
ax1.text(0, f1(0), f'(0, 0)', fontsize=12, verticalalignment='bottom')
ax1.set_title('Function: $f(x) = |x|$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.axhline(0, color='black', linewidth=0.5)
ax1.axvline(0, color='black', linewidth=0.5)
ax1.legend()
ax1.grid(True)
ax2.plot(x, f2(x), label=r'$g(x) = x^{1/3}$', color='blue')
ax2.scatter([0], [f2(0)], color='green', zorder=5)
ax2.text(0, f2(0), f'(0, 0)', fontsize=12, verticalalignment='bottom')
ax2.set_title(r'Function: $g(x) = x^{1/3}$')
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.axhline(0, color='black', linewidth=0.5)
ax2.axvline(0, color='black', linewidth=0.5)
ax2.legend()
ax2.grid(True)
plt.tight_layout()
plt.show()
```

1. in the case of $f(x) = \mid x \mid$
$$
f'(x) = \lim_{x \to 0} \frac{\mid x \mid - 0}{x - 0} = \frac{\mid x \mid}{x}
$$
let's break the limit down into left and right
$$
\lim_{x \ \to 0^{+}} \frac{\mid x \mid}{x} = 1
$$
$$
\lim_{x \ \to 0^{-}} \frac{\mid x \mid}{x} = -1
$$
so the $f'(0) DNE$ 

2. in the case of $g(x) = x^{\frac{1}{3}}$
$$
g'(x) = \lim_{x \to 0} \frac{x^\frac{1}{3} - 0}{x - 0} = \pm \infty
$$

so the function has a vertical tangent line, and the slope of a vertical line is undefined,

thus $g'(0) DNE$

## Differentiation rules

$$
\begin{gather*}
    \frac{d}{dx} [C] = 0 \\
    \\
    \frac{d}{dx} [x^C] = Cx^{(c-1)} \\
    \\
    (f+g)' = f' + g' \\
    \\
    (fg)' = f'g + fg' \\
    \\
    cf = cf' \\
    \\
    (\frac{f}{g})' = \frac{f'g - g'f}{g^2}
\end{gather*}
$$

### chain rule

let $f$ and $g$ be functions defined at and around $x$
* IF: $f$ is Differentiable at $g(x)$, $g$ is Differentiable at $x$
* THEN: $(f \circ g)$ is Differentiable at $x$ and
$$
(f \circ g)'(x) = f'(g(x)).g'(x)
$$

### trigonometric functions

$$
\begin{gather*}
    \frac{d}{dx}[\sin x] = \cos x \\
    \\
    \frac{d}{dx} [\cos x] = -\sin x \\
    \\
    \frac{d}{dx} [\tan x] = \sec^{2} x \\
    \\
    \frac{d}{dx} [\cot x] = -\csc^2 x \\
    \\
    \frac{d}{dx} [\sec x] = \sec x\tan x \\
    \\
    \frac{d}{dx} [\csc x] = - \csc x\cot x
\end{gather*}
$$

### some proofs

let's prove some of the Differentiation rules

Ex1) prove $(fg)' = f'g + fg'$

* $f$ and $g$ are differentiable function at and around $a \in \mathbb{r}$ and $h = fg$

pf:
$$
h'(a) = \lim_{x \to a} \frac{h(x) - h(a)}{x-a} = \lim_{x \to a} \frac{f(x)g(x) - f(a)g(a)}{x-a} 
$$
$$
= \lim_{x \to a} \frac{f(x)g(x) - f(a)g(a)\pm f(a)g(x)}{x-a} = \lim_{x \to a} [\frac{f(x) - f(a)}{x-a}.g(x) + \frac{g(x) - g(a)}{x-a}.f(a)]
$$
- ($f$ and $g$ are difrentiable) and thus ($g$ is continuous at $a$) 
$$
= f'(a).\lim_{x \to a} g(x) + g'(a)f(a) = f'(a)g(a) + f(a)g'(a)
$$

We have shown $(fg)' = f'g + fg'$, as needed $\square$

Ex2) prove  $\frac{d}{dx} [x^c] = cx^{c-1}$  ($c \in {1,2,3,...}$)

pf: (Proof by induction on $c$)

Base Case: ($c =1$)
$$
\frac{d}{dx} x^c = \lim_{h \to 0} \frac{(x+h)^c - x^c}{h} =(c=1) \lim_{h \to 0} \frac{(x+h)^1 - x^1}{h} = \lim_{h \to 0} \frac{h+x-x}{h} =\lim_{h \to 0} \frac{h}{h} = 1
$$

Induction step: $\frac{d}{dx} [x^c] = cx^{c-1}$ $\Rightarrow$ $\frac{d}{dx} [x^{c+1}] = (c+1)x^{c}$
* Assume(hyp.) $\frac{d}{dx} [x^c] = cx^{c-1}$, WTS:$\frac{d}{dx} [x^{c+1}] = (c+1)x^{c}$

$$
\frac{d}{dx} [x^{c+1}] = \frac{d}{dx} [x^{c} . x] = ($(fg)' = f'g + fg'$) \frac{d}{dx}[(x^{c})].x + x^{c} .\frac{d}{dx} x
$$

$$
= (hyp) cx^{c-1}.x + x^{c}.1 =  cx^c + x^c = (c+1)x^c
$$
we have shown $\frac{d}{dx} [x^{c+1}] = (c+1)x^{c}$, as needed $\square$

### Implicit differentiation

Problem: Find the slope of the line tangent to the graph of $3x^2 + 2y^2 - 4xy + 2x - 8y + 11 = 0$ at $(1,4)$.

**Think of $y = f(x)$ Implicitly**

IF $3x^2 + 2y^2 - 4xy + 2x - 8y + 11 = 0$

THEN $\frac{d}{dx} [3x^2 + 2y^2 - 4xy + 2x - 8y + 11] = \frac{d}{dx} [0]$

let's calculate:
$$
\frac{d}{dx} [ 3x^2 + 2y^2 - 4xy + 2x - 8y + 11] = 6x + 4yy' - (4y + 4xy') + 2 - 8y' + 0 = \frac{d}{dx} [0] = 0
$$
When $x=1$ and $y=4$
$$
 6x + 4yy' - (4y + 4xy') + 2 - 8y' + 0 = 6 +16y' - 16 - 4y' + 2 - 8y' = 0 \rightarrow -8y' = -4 \rightarrow y' = 2
$$
