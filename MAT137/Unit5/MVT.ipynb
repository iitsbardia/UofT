# The Mean Value Theorem and its applications.

- $f$ has a local maximum at $c$ when
$$
\exists \delta > 0 \quad s.t. \quad |x-c| > 0 \implies f(x) \leq f(c)
$$

### Local EVT

Let $f$ be a function on $I$

IF
- $f$ has a local extrema at $c$, and
- $c$ is not an endpoint

Then
$$
f'(c) = 0 \quad \text{ or }\quad f'(c) \text{ DNE}
$$

pf:
$$
f'(c) = \lim_{x\to c}\frac{f(x) - f(c)}{x-c}
$$
$$
\begin{cases}
\lim_{x\to c^{+}}\frac{f(x) - f(c)}{x-c} \geq 0 \\
\lim_{x\to c^{-}}\frac{f(x) - f(c)}{x-c} \leq 0
\end{cases}
$$
thus, if $f'(c)$ exists it is $0$ $\square$

---

Ex: Find extrema of $f$ on $[-4,4]$
$$
f(x) = x^3 - 3x^2 - 9x + 35
$$

Solution:
Since $f$ is continuous on $[-4,4]$ then by EVT, it has a minimum and a maximum!
$$
f'(x) = 3x^2 - 6x -9 = 3(x+1)(x-3)
$$
| x | type | f(x) |  |
|:------:|:------:|:------------------:|:------------:|
| -4      |   endpoint   | -41                  | MIN           | 
| -1    | critical point |40                  | MAX / local max          | 
| 3     |   critical point   | 8                  | local min            | 
| 4      | endpoint      | 15                  | T            | 

$x = -4$ is the minimum and $x = -1$ is the maximum of $f$

### Rolle's theorem

Let $a<b$, let $f$ be a function defined on $[a,b]$
IF
- $f$ is difrentiable on $(a,b)$
- $f$ is continuous on $[a,b]$
- $f(a) = f(b)$

THEN
$$
\exists c \in (a,b) \qquad s.t. \qquad f'(c) = 0
$$

PF:

Since $f$ is continuous on $[a,b]$ then by EVT it should have a max and min on $[a,b]$

* Case(1):

$f$ has a max or min at $c \in$ $(a,b)$, then by local EVT $f'(c) = 0$ or DNE, since $f$ is difrentiable on $(a,b)$, thus $f'(c) = 0$

* Case(2):

$f$ has both max and min at endpoints, then the function is a constant and there are infinitely $c \in (a,b)$ s.t. $f'(c) = 0$ 

We have shown in all cases there is at least one $c \in (a,b)$ s.t. $f'(c) =0$, as needed $\square$

## MVT

Let $f$ be a function on an interval $[a,b]$
IF

$f$ is continuous on $[a,b]$ and $f$ is difrentiable on $(a,b)$

THEN

$$
\exists c \in (a,b) \quad s.t. \quad f'(c) = \frac{f(b) - f(a)}{b-a}
$$

PF:

Let $m =  \frac{f(b) - f(a)}{b-a}$ , Let $h$ be a function on $[a,b]$ s.t. $ h(x) = f(x) - f(a) - m(x-a)$

Since $f$ is continuous on $[a,b]$ and $f$ is difrentiable on $(a,b)$, Then $h$ is as well

$h'(x) = f'(x) - m$

since $h(a) = 0$ and $h(b) = 0$ thus by Rolle's theorem $ \exists c \in (a,b) \quad s.t. \quad h'(c) = 0$

Thus  $\exists c \in (a,b) \quad s.t. \quad h'(c) = f'(c) - m  = 0$, We have shown $\exists c \in (a,b) \quad s.t. \quad f'(c) = m$ as needed $\square$

## MVT applications

---

**Claim**

If $f' = 0$ on $I$ then $f$ is a constant on $I$

**Theorem**

Let $f$ be a function on $[a,b]$

IF
$$
\forall x \in (a,b), \quad f'(x) = 0
$$
THEN
$$
f \text{ is a constant on } [a,b]
$$

PF:

WTS : $\forall x_{1}, x_{2} \in [a,b], \quad f(x_{1}) = f(x_{2})$

Fix $x_{1}, x_{2} \in [a,b]$, assume $x_{1} < x_{2}$
Since $[x_{1}, x_{2}] \subseteq [a,b]$ $f$ is continuous and differentiable on them

Hence, by MVT
$$
\exists c \in (x_{1}, x_{2}) \quad s.t. \quad f'(c) = \frac{f(x_{2}) - f(x_{1})}{x_{2} - x_{1}}
$$
Since $f'(c) = 0$, thus we get $f(x_{1}) = f(x_{2})$ as needed $\square$

---

**Theorem**

IF
$$
\forall x \in \mathbb{R}, \quad f'(x) = x^2
$$
THEN
$$
\forall x \in \mathbb{R}, \quad f(x) = \frac{1}{3} x^3 + C
$$

PF:
$$
\frac{d}{dx}[f(x)] = \frac{d}{dx}[\frac{1}{3} x^3]
$$
Thus, (since they have the same derivative, their sub is a constant) $f(x) - \frac{1}{3} x^3 = C$ $\square$
