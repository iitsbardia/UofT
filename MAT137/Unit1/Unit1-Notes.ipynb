# Unit 1: Logic, sets, notation, definitions, and proofs
### Sets and notation
important sets of numbers:
* Naturals  $\mathbb{N} = \{0,1,2,3,\ldots\}$
* Integers  $\mathbb{Z} = \{\ldots,-3,-2,-10,1,2,3,\ldots\}$
* Rationals  $\mathbb{Q} = \{\frac{n}{m},\forall n,m \in \mathbb{z}\}$
* Real  $\mathbb{R}$ = {rational and irrational numbers}

### Set-building notation
There are many way to anotate a set
1. one way is to define a range for x and set the conditions
$$
A = \{ x \in \mathbb{Z} : x^2 < 6\}  
$$
$$
A = \{-2,-1,0,1,2\}
$$
2. another way is as follows
$$
B = \{ 2x : x \in A \}  
$$
$$
B = \{-4,-2,0,2,3\}
$$
3. another way is using intervals
$$
[a,b] = \{x \in \mathbb{R} : x>=a , x<=b\}
$$
$$
(a,b) = \{x \in \mathbb{R} : x>a , x<b\}
$$
$$
[a,\infty) = \{x \in \mathbb{R} : x>=a\}
$$
$$
(-\infty,b) = \{x \in \mathbb{R} : x<=b\}
$$

### Quantifiers
|   Syntax  | Description |
| ----------- | ------------------ |
| $\forall$     | For all / Every      |
| $\exists$  | There exist / is       |

Ex:
1. &nbsp;$\forall x \in \mathbb{R} , x^2 >= 0$ &nbsp;&nbsp;  TRUE
1. &nbsp;$\forall x \in \mathbb{Z} , x>\pi$ &nbsp;&nbsp;  FALSE
1. &nbsp;$\exists x \in \mathbb{R}$ &nbsp; $\mathcal{s}.\mathcal{t}$ &nbsp; $x^2 = 5$ &nbsp;&nbsp;  TRUE
1. &nbsp;$\exists x \in \mathbb{R}$ &nbsp; $\mathcal{s}.\mathcal{t}$ &nbsp; $x^2 = -1$ &nbsp;&nbsp;  FALSE

### Double quantifiers

take these two examples:
1. $\forall x \in \mathbb{Z}$ , $\exists y \in \mathbb{Z}$&nbsp; $\mathcal{s}.\mathcal{t}$ &nbsp;$x<y$
* This statement is TRUE, it reads "Every integer is smaller than some other one."
3. $\exists x \in \mathbb{Z}$ &nbsp; $\mathcal{s}.\mathcal{t}$ &nbsp; $\forall y \in \mathbb{Z}$ , $x<y$
* This statement is FALSE, it reads "There exists an integer smaller than all other one."

**A simple proof**  
Let's prove the first statement above, we already have established that it is true, so let's write a proof for it  
WTS : $\forall x \in \mathbb{Z}$ , $\exists y \in \mathbb{Z}$&nbsp; $\mathcal{s}.\mathcal{t}$ &nbsp;$x<y$  
proof:  
* Let $x \in \mathbb{Z}$
* Take $y = x+1$
* Verify:
- [x] $y \in \mathbb{Z}$
- [x] $y = x+1 > x $
$\square$

### Conditional statements

in a conditional statement we say whenever p is True then q must be True ass well  
we notate that statement like this:
1. if $P$, then $Q$
1. $P \Rightarrow  Q$
2. $P$ implies $Q$

The values of a conditional statement is as follows
| P     | Q | P implies Q    |
| :---        |    :---:   |          -------: |
| T    | T       | T   |
| T    | F       | F      |
| F    | T       | T              |
| F    | F       | T              |
* the statement is only False when P is True but Q is False

Whenever we have &nbsp; "$P \Rightarrow  Q$" &nbsp; we also have &nbsp;"$not P$&nbsp;$ \Rightarrow $&nbsp; $not Q$"  
These two statements mean  **Exactly the same thing**  

When $P \Rightarrow  Q$ and $Q \Rightarrow  P$ we can say  &nbsp; $Q \iff  P$  
this statement read **"$P$ if and only if $Q$"** or **"$P$ iff $Q$"** for short  
In this scenario $P$ and $Q$ are equivalent or $P$ and $Q$ are both either True or False

#### Negation of a Conditional statement
Ex:  
* Let $A \subset \mathbb{R}$  
$ x \in A \Rightarrow x>0 $  
the negation of this statement is:  
$\exists x \in \mathbb{R}$ &nbsp;$s.t$ &nbsp;($ x \in A$ &nbsp;&nbsp;$\wedge$ &nbsp;&nbsp;$x \leq 0$)

## Proofs

Let's define and increase a function, we say a function is increasing on $I$ when:  
$\forall x_{1},x_{2} \in I$,&nbsp; $x_{1}>x_{2} \Rightarrow f(x_{1}) > f(x_{2})$  

Now let's try to write a proof from the definition  
Ex: Prove $f(x) = 3x + 7$ is increasing on $\mathbb{R}$ directly from the definition  
Pf:
* WTS : $\forall x_{1},x_{2} \in \mathbb{R}$,&nbsp; $x_{1}>x_{2} \Rightarrow f(x_{1}) > f(x_{2})$ 
* Let $x_{1},x_{2} \in \mathbb{R}$
* assume $x_{1}>x_{2}$, we need to confirm $f(x_{1})>f(x_{2})$
* Since $x_{1}>x_{2}$, &nbsp; &nbsp; $f(x_{1}) = 3x_{1} + 7 > 3x_{2} +7 = f(x_{2})$
* thus, we have shown $f(x_{1})>f(x_{2})$ as needed.$\square$

Now let's write a proof to show a statement is False  
Ex: Prove $g(x) = \cos x$ is not increasing on [0,$\pi$] directly from definition  
Pf:  
* We need to prove that the negation of the definition function is increasing is True for $g(x)$
* WTS : $\exists x_{1},x_{2} \in [0,\pi]$&nbsp; $\mathcal{s}.\mathcal{t}$ &nbsp;&nbsp; ($x_{1}>x_{2}$&nbsp; $\wedge$&nbsp; $g(x_{1}) < g(x_{2})$)
* take $x_{1} = \pi$ , $x_{2} = \frac{\pi}{2}$
* Verify,$x_{1} > x_{2}$✓ and $ -1 = g(x_{1}) < g(x_{2}) = 0$✓ $\square$

Let's prove a theorem  
Ex: if $f$ and $g$ are function and increasing on $I$ and $h = f + g$ then $h$ is also increasing on $I$  
Pf:
* let $f(x)$ and $g(x)$ be increasing function on $I$ and $h(x)$ = $f(x)$ + $g(x)$
* WTS: $\forall x_{1},x_{2} \in I$, $x_{1} > x_{2}$ $\Rightarrow$ $h(x_{1}) > h(x_{2})$
* let $x_{1},x_{2} \in I$
* Assume $x_{1} > x_{2}$
* Since $x_{1} > x_{2}$ and $f(x)$ is on $I$ $\Rightarrow$ $f(x_{1}) > f(x_{2})$
* Since $x_{1} > x_{2}$ and $g(x)$ is on $I$ $\Rightarrow$ $g(x_{1}) > g(x_{2})$
* thus, $h(x_{1}) = g(x_{1}) + f(x_{1})  > f(x_{2}) + g(x_{2})= h(x_{2})$ $\square$

Proof by Induction

Mathematical induction is a method for proving that a statement $S(n)$ is true for every $n\in \mathbb{N}$  
The proof consists of two steps:
1. Base case: proving $S(n)$ for smallest value on $n$
2. Induction step: Proving that $S(n)$ $\Rightarrow$ $s(n+1)$

Ex: Prove for every $n \geq 4$, $n! > 2^n$

Pf: (Proof by induction)
* take $n \geq 4$
* Base case($n = 4$):
    - $4! = 24$ > $16 = 2^4$ ✓
* Induction step:
    - WTS: $n! > 2^n$ $\rightarrow$ $(n+1)! > 2^{n+1}$
    - assume $n! > 2^n$ **(induction hypothesis)**
    - Since $n \geq 4$, we know $ n+1 > 2$
    - Now since every element is none-negative we can multiply and get: $n! (n+1)$ $\geq$ $2^n.2$
    - thus we have shown, $n! > 2^n$ $\rightarrow$ $(n+1)! > 2^{n+1}$ as needed. $\square$
